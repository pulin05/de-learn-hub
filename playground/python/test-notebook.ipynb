```python
numbers = "10, 20, 30, 40"

for x in numbers.split():
    print(x)
```

```python
def insert_at_middle(arr, index, value):
    arr.append(None)
    for i in range(len(arr) -1, index, -1):
        arr[i] = arr[i-1]
    arr[index] = value
    return arr

numbers = [10, 20, 30, 40]

print(insert_at_middle(numbers, 2, 25))
```
